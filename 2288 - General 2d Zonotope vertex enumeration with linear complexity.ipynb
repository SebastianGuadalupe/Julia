```julia
using Revise, LazySets, BenchmarkTools, Plots
```

## Yupbank

```julia
bench = []
for i in 1:20
    Z = rand(Zonotope, num_generators=i, dim=2)
    t = @benchmark LazySets._vertices_list_2D($Z.center, $Z.generators, apply_convex_hull=false) samples=10
    t = minimum(t).time
    push!(bench, t)
end
```

```julia
plot(bench)
```

## Master

```julia
bench2 = []
for i in 1:20
    Z = rand(Zonotope, num_generators=i, dim=2)
    t = @benchmark LazySets._vertices_list_2D($Z.center, $Z.generators, apply_convex_hull=false) samples=10
    t = minimum(t).time
    push!(bench2, t)
end
```

```julia
plot(bench, label="yupbank")
plot!(bench2, label="master")
```

```julia
plot(bench[1:10], label="yupbank")
plot!(bench2[1:10], label="master")
```
